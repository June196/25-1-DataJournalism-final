```python
# 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import re

# 데이터 불러오기
df = pd.read_csv('CourseList.csv')

# 한글 폰트 경로 직접 지정
korean_font = './NanumGothic.ttf'

# 워드클라우드 생성 함수 정의
def generate_wordcloud_by_area(df, font_path):
    unique_areas = df['영역'].dropna().unique()
    n_cols = 2
    n_rows = (len(unique_areas) + 1) // n_cols

    plt.figure(figsize=(14, 4 * n_rows))

    for i, area in enumerate(unique_areas, 1):
        area_df = df[df['영역'] == area]
        text = " ".join(area_df['교과목명'].astype(str))

        wordcloud = WordCloud(
            font_path=font_path,
            width=800,
            height=400,
            background_color='white'
        ).generate(text)

        plt.subplot(n_rows, n_cols, i)
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title(f"{area} 영역 WordCloud")

    plt.tight_layout()
    plt.show()

generate_wordcloud_by_area(df, font_path=korean_font)
```

```python
# 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 불러오기
df = pd.read_csv('CourseRequirement.csv')

# 분석 대상 영역
area_cols = ['학문의 토대', '지성의 열쇠', '베리타스', '지성의 확장']

# 영역별 비율 계산
ratio_df = df.copy()
ratio_df[area_cols] = ratio_df[area_cols].div(ratio_df['최소 학점'], axis=0)

# 시각화를 위한 전처리 (인덱스를 대학으로 설정)
ratio_df.set_index('대학', inplace=True)

# 누적 막대그래프
plt.figure(figsize=(12, 6))
bottom = None
for col in area_cols:
    plt.bar(ratio_df.index, ratio_df[col], label=col, bottom=bottom)
    bottom = ratio_df[col] if bottom is None else bottom + ratio_df[col]

plt.title('대학별 공통교육 영역 구성 비율 (누적 막대그래프)')
plt.ylabel('비율')
plt.xticks(rotation=45, ha='right')
plt.legend(title='공통교육 영역', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
